```python
import pickle
import pandas as pd

# Re-train Linear Regression model to ensure it's available for evaluation
# (This assumes X_train, y_train are available from previous steps)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
print('Linear Regression model re-trained successfully.')

# Re-evaluate all models to get an up-to-date metrics_df
# (This code block is adapted from cell 36f83f8c)
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

model_names = []
r2_scores = []
maes = []
mses = []

# Evaluate Linear Regression model
y_pred_lr = linear_model.predict(X_test)
model_names.append('Linear Regression')
r2_scores.append(r2_score(y_test, y_pred_lr))
maes.append(mean_absolute_error(y_test, y_pred_lr))
mses.append(mean_squared_error(y_test, y_pred_lr))

# Evaluate Decision Tree Regressor model
if 'dt_model' in locals(): # Ensure dt_model exists from previous training
    y_pred_dt = dt_model.predict(X_test)
    model_names.append('Decision Tree Regressor')
    r2_scores.append(r2_score(y_test, y_pred_dt))
    maes.append(mean_absolute_error(y_test, y_pred_dt))
    mses.append(mean_squared_error(y_test, y_pred_dt))

# Evaluate Random Forest Regressor model
if 'rf_model' in locals(): # Ensure rf_model exists from previous training
    y_pred_rf = rf_model.predict(X_test)
    model_names.append('Random Forest Regressor')
    r2_scores.append(r2_score(y_test, y_pred_rf))
    maes.append(mean_absolute_error(y_test, y_pred_rf))
    mses.append(mean_squared_error(y_test, y_pred_rf))

# Evaluate K-Nearest Neighbors Regressor model
if 'knn_model' in locals(): # Ensure knn_model exists from previous training
    y_pred_knn = knn_model.predict(X_test)
    model_names.append('K-Nearest Neighbors Regressor')
    r2_scores.append(r2_score(y_test, y_pred_knn))
    maes.append(mean_absolute_error(y_test, y_pred_knn))
mses.append(mean_squared_error(y_test, y_pred_knn))

# Evaluate Support Vector Machine Regressor model
if 'svm_model' in locals(): # Ensure svm_model exists from previous training
    y_pred_svm = svm_model.predict(X_test)
    model_names.append('Support Vector Machine Regressor')
    r2_scores.append(r2_score(y_test, y_pred_svm))
    maes.append(mean_absolute_error(y_test, y_pred_svm))
    mses.append(mean_squared_error(y_test, y_pred_svm))

metrics_df = pd.DataFrame({
    'Model': model_names,
    'R2 Score': r2_scores,
    'MAE': maes,
    'MSE': mses
})

display(metrics_df.sort_values(by='R2 Score', ascending=False))

# Identify the best model based on R2 Score
best_model_row = metrics_df.loc[metrics_df['R2 Score'].idxmax()]
best_model_name = best_model_row['Model']

# Get the actual best model object
best_model = None
if best_model_name == 'Linear Regression':
    best_model = linear_model
elif best_model_name == 'Decision Tree Regressor':
    best_model = dt_model
elif best_model_name == 'Random Forest Regressor':
    best_model = rf_model
elif best_model_name == 'K-Nearest Neighbors Regressor':
    best_model = knn_model
elif best_model_name == 'Support Vector Machine Regressor':
    best_model = svm_model

if best_model is not None:
    # Save the best model to a .pkl file
    filename = f'{best_model_name.replace(" ", "_").lower()}_model.pkl'
    with open(filename, 'wb') as file:
        pickle.dump(best_model, file)
    print(f'Best model ({best_model_name}) saved successfully to {filename}')
else:
    print('Could not identify or save the best model.')
```
